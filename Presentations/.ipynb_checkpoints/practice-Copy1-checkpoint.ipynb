```python
def rename_col(df):
    df.columns = df.iloc[0]
    df = df.drop([0])
    return df

def field_col(df,col,ref):
    cond = df[col].isna()
    df['Field'] = np.where(cond,df[ref],np.nan)
    return df

def fill_col(df,col):
    df[col] = df[col].ffill()
    return df

def remove_rows(df):
    cond = df.iloc[:,0].eq(df.iloc[:,-1])
    df = df.loc[~cond]
    return df
    

(pd
 .read_excel('worked-examples.xlsx',
             sheet_name='pivot-hierarchy',
             header=None)
 .remove_empty()
 .pipe(field_col,2,1)
 .pipe(fill_col,'Field')
 .pipe(remove_rows)
 .replace({1:{np.nan:'Subject'},
           'Field':{np.nan:'Field'}})
 .pipe(rename_col)
 .melt(id_vars=['Subject','Field'],value_name='Score',var_name='Student')

)
```

```python
#requires openpyxl to determine border locations

wb = load_workbook(filename='worked-examples.xlsx')

ws = wb['pivot-centre-aligned']

box = ws['B2':'J11']

cols = set()
rows = set()

for i in box:
    for cell in i:
        if cell.border.right.style:
            cols.add(cell.column)
        elif cell.border.bottom.style:
            rows.add(cell.row)
            
cols = sorted(cols)
rows = sorted(rows)
rows = list(windowed(rows,2))
cols = list(windowed(cols,2))




df = (pd
 .read_excel('worked-examples.xlsx',
             sheet_name='pivot-centre-aligned',
             header=None)

)

df

gender_and_names = [df.copy()
                    .iloc[:,start:end]
                    .dropna(how='all')
                    .ffill(axis=1)
                    .bfill(axis=1)
                    for start,end
                    in cols
                   ]

gender = pd.concat(gender_and_names,axis=1)

Fields = [df.copy()
          .iloc[start:end]
          .dropna(how='all',axis=1)
          .ffill()
          .bfill()
          .iloc[:,:2]
          for start,end in rows
         ]

fields = pd.concat(Fields)

def rename_col(df):
    df.columns = df.iloc[:2].add(',').sum().str.strip(',')
    df = df.iloc[2:]
    return df

def create_col(df,col,new_col):
    df[new_col] = df[col].str.split(',').str[0]
    df[col] = df[col].str.split(',').str[-1]
    return df

(pd.concat([fields,gender],axis=1)
 .fillna(value={1:'Field',
                2:'Subject'},
         limit=1)
 .fillna('')
 .pipe(rename_col)
 .melt(id_vars=['Field','Subject'], var_name='Student', value_name='Score')
 .pipe(create_col,'Student','Gender')
)
```

```python
def remove_rows(df,col,cond):
    cond = df[col].str.startswith(cond,na=False)
    df = df.loc[~cond]
    return df

(pd
 .read_excel('worked-examples.xlsx',
             sheet_name='pivot-repeated-headers',
             header=None)
 .remove_empty()
 .ffill()
 .replace({1:{np.nan:'Course'},
           2:{np.nan:'Student'}
          })
 .row_to_names(row_number = 0,
               remove_row = True)
 .melt(id_vars=['Course','Student'],
       value_name='Score',
       var_name='Term')
 .pipe(remove_rows,'Score','Term')
 .reset_index(drop = True)
)
```

```python
def create_column_null(df,col,ref,new_col):
    cond = df[col].isna()
    df[new_col] = np.where(cond,df[ref],np.nan)
    return df

def fill_null(df,col):
    df[col] = df[col].ffill()
    return df

def extract_col(df,col,new_col):
    df[new_col] = df[col].str.extract(r'(Term\s\d)')
    return df

(pd
 .read_excel('worked-examples.xlsx',
             sheet_name='pivot-header-within-data',
             header=None)
 .remove_empty()
 .pipe(create_column_null,3,2,'Field')
 .pipe(fill_null,'Field')
 .rename(columns={1:'Student'})
 .melt(id_vars=['Field','Student'],value_name='Score')
 .pipe(extract_col,'Score','Term')
 .pipe(fill_null,'Term')
 .dropna(how='any')
 .drop('variable',axis=1)
 .reset_index(drop=True)
)
```

```python
def row_indices(df):
    row_filter = df.isin(['Name']).any(axis=1)
    ind = [index -1
           for index in
           df.loc[row_filter].index]
    ind.append(len(df))
    ind = list(windowed(ind,2))
    return ind
    
def col_indices(df):
    col_filter = df.isin(['Name','Grade']).any()
    col = [df.columns.get_loc(ind) 
           for ind in 
           df.loc[:,col_filter].columns]
    col = list(chunked(col,2))
    return col

def extract_col(df,col,ref,new_col):
    cond = df[col].isna()
    df[new_col] = np.where(cond,df[ref],np.nan)
    return df

def remove_rows(df,col,ref):
    cond = df[col] != df[ref]
    df = df.loc[cond]
    return df




df = (pd
      .read_excel('worked-examples.xlsx',
                  sheet_name='small-multiples',
                  header=None)
      .remove_empty()
     )

cols = col_indices(df)
rows = row_indices(df)

outcome = [df.copy()
           .iloc[start:finish,begin:end+1]
           .reset_index(drop=True)
           .row_to_names(row_number=1,remove_row=True)
           for start,finish in rows
           for begin,end in cols]

merger = (pd.concat(outcome)
          .pipe(extract_col,'Score','Name','Field')
          .ffill()
          .pipe(remove_rows,'Name','Field')
          .reset_index(drop=True)
         )

merger
```

```python
def process(df):
    outcome = (df.copy()
               .melt(id_vars=['Unnamed: 0'],value_name='Score', var_name='Student')
               .rename(columns={'Unnamed: 0':'Student'}) 
              )
    
    return outcome
```

```python
data_dict = (pd.read_excel('worked-examples.xlsx',
                           sheet_name=['humanities','performance']
                          )
             .items()
            )

_ = [process(data)
     .assign(field=field)
     for field, data 
     in data_dict
    ]

pd.concat(_,ignore_index=True)
```

```python
def filter_df(df,text,col):
    cond = df[col]==text
    ind = df.loc[cond].index[0]
    df = df.loc[ind:]
    return df


    

def process(df):
    df = (df
          .remove_empty()
          .pipe(filter_df,'Subject',0)
          .row_to_names(0,True)
          .melt(id_vars='Subject',
                var_name='Student',
                value_name='Score')
            )
    
    return df


data_dict = (pd.read_excel('worked-examples.xlsx',
                          sheet_name=['female','male'],
                          header=None)
             .items()
            )


output = [process(data)
          .assign(gender=gender)
          for gender,data
          in data_dict]

pd.concat(output,ignore_index=True)
```

```python
def rename_cols(df):
    
    df.columns = (df
                  .iloc[:2]
                  .add(',')
                  .sum()
                  .str.strip(',')
                 )
    
    df = df.drop([0,1])
    
    return df


def extract_col_grade(df,col,new_col,ref,text):
    
    cond = df[col]==text
    
    df[new_col] = np.where(cond,df[ref],np.nan)
    
    return df


(pd.read_excel('worked-examples.xlsx',
               sheet_name='implied-multiples',
               header=None)
 .ffill(axis=1)
 .T
 .fillna('Field')
 .row_to_names(0,True)
 .melt(id_vars=['Field','Name'], var_name='Student', value_name = 'Score')
 .pipe(extract_col_grade,'Name','Grade','Score','Grade')
 .bfill()
 .query('Name != "Grade"')
 .reset_index(drop = True)
 .rename(columns={"Name":'Subject'})
)
```
